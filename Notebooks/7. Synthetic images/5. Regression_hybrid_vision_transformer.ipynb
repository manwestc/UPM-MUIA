# <font color="#004D7F" size=6> 1. Description</font>

The growing interest in the use of algorithms-based machine learning for predictive tasks has generated a large and diverse development of algorithms. However, it is widely known that not all of these algorithms are adapted to efficient solutions in certain tidy data format datasets. For this reason, novel techniques are currently being developed to convert tidy data into images with the aim of using Convolutional Neural Networks (CNNs) or Vision Transformer (ViT). TINTOlib offers the opportunity to convert tidy data into images through several techniques: TINTO, IGTD, REFINED, SuperTML, BarGraph, DistanceMatrix and Combination.

# <font color="#004D7F" size=5> 1.1. Main Features</font>

- Supports all CSV data in **[Tidy Data](https://www.jstatsoft.org/article/view/v059i10)** format.
- For now, the algorithm converts tabular data for binary and multi-class classification problems into machine learning.
- Input data formats:
    - **Tabular files**: The input data could be in **[CSV](https://en.wikipedia.org/wiki/Comma-separated_values)**, taking into account the **[Tidy Data](https://www.jstatsoft.org/article/view/v059i10)** format.
    - **Dataframe***: The input data could be in **[Pandas Dataframe](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html)**, taking into account the **[Tidy Data](https://www.jstatsoft.org/article/view/v059i10)** format.
    - **Tidy Data**: The **target** (variable to be predicted) should be set as the last column of the dataset. Therefore, the first columns will be the features.
    - All data must be in numerical form. TINTOlib does not accept data in string or any other non-numeric format.
- Runs on **Linux**, **Windows** and **macOS** systems.
- Compatible with **[Python](https://www.python.org/)** 3.7 or higher.

This is an example of a simple ViT for TINTOlib images. Note that we are not looking for the optimization of the ViT but to show an example of TINTOlib execution.

It is crucial to select an appropriate patch size. The patch size should be a divisor of the input image size; for example, an image of 20x20 with a patch size of 5 would result in a total of 16 patches (4x4 grid). Given the high computational cost, the patch size should be carefully chosen based on the dimensions of the image.

Este código define los hiperparámetros para el modelo ViT implementado en TensorFlow y Keras. Los hiperparámetros son los siguientes:

* `learning_rate`: tasa de aprendizaje para el optimizador.
* `weight_decay`: valor de decaimiento de peso para regularización L2.
* `batch_size`: tamaño del lote de entrenamiento.
* `num_epochs`: número de épocas de entrenamiento.
* `image_size`: tamaño de las imágenes de entrada.
* `patch_size`: tamaño de los parches extraídos de las imágenes.
* `num_patches`: número total de parches extraídos de cada imagen.
* `projection_dim`: dimensión de la proyección lineal para los parches.
* `num_heads`: número de cabezas de atención en el transformador.
* `transformer_units`: lista de unidades en las capas del transformador.
* `transformer_layers`: número de capas en el transformador.
* `mlp_head_units`: lista de unidades en las capas densas del clasificador final.

Estos hiperparámetros se utilizan para configurar el modelo ViT y su proceso de entrenamiento.

<a id="section521"></a>
## <font color="#004D7F" size=5> 5.2.1. Clase `Patches`</font>

La clase `Patches` divide una imagen en pequeños parches de tamaño fijo, reorganizándolos en un tensor que puede ser utilizado como entrada para un Vision Transformer. Esto es esencial porque los Transformers trabajan con secuencias, y esta clase convierte las imágenes en secuencias de parches.

<figure><center>
  <img src="../../img/patch_embedding.png" width="850" height="200" alt="Gráfica">
  <figcaption><blockquote>Arquitectura de ViT - Patch embedding. Extraída de <a href="https://arxiv.org/abs/2010.11929">Aman Arora's Blog</a></blockquote></figcaption>
</center></figure>

La clase `Patches` es una subclase de `layers.Layer` en Keras, lo que significa que es una capa personalizada. Esta clase se encarga de dividir una imagen en pequeños parches (patches) que serán utilizados como entradas en el Vision Transformer.

##### Constructor `__init__`
- `__init__` es el constructor de la clase.
- `patch_size` es un parámetro que especifica el tamaño de cada parche (patch) en que se dividirá la imagen.
- `super().__init__()` llama al constructor de la clase base (`layers.Layer`), inicializando la capa.

##### Método `call`
- `call` es el método que define la lógica de la capa. Es invocado cuando se pasa un tensor (en este caso, imágenes) a través de la capa.
- `input_shape = ops.shape(images)` obtiene la forma (dimensiones) del tensor de entrada `images`. Asumiendo que `images` es un tensor 4D (batch, height, width, channels).
- `batch_size`, `height`, `width`, y `channels` extraen las respectivas dimensiones de la imagen.
- `num_patches_h` y `num_patches_w` calculan el número de parches en la altura y el ancho de la imagen, respectivamente, dividiendo la dimensión correspondiente por `patch_size`.
- `patches = keras.ops.image.extract_patches(images, size=self.patch_size)` utiliza una función de Keras para extraer parches de tamaño `patch_size` de las imágenes de entrada.
- `patches = ops.reshape(...)` reconfigura el tensor `patches` para tener la forma `(batch_size, num_patches_h * num_patches_w, patch_size * patch_size * channels)`. Esto significa que cada parche se aplana y se organiza en una secuencia de parches.

##### Método `get_config`

- `get_config` es un método estándar en Keras para capas personalizadas. Permite que la configuración de la capa se pueda guardar y volver a cargar.
- `config = super().get_config()` llama al método `get_config` de la clase base para obtener la configuración básica de la capa.
- `config.update({"patch_size": self.patch_size})` añade el `patch_size` a la configuración.
- `return config` devuelve la configuración completa.

````python
class Patches(layers.Layer):
    def __init__(self, patch_size):
        super().__init__()
        self.patch_size = patch_size

    def call(self, images):
        input_shape = ops.shape(images)
        batch_size = input_shape[0]
        height = input_shape[1]
        width = input_shape[2]
        channels = input_shape[3]
        num_patches_h = height // self.patch_size
        num_patches_w = width // self.patch_size
        patches = keras.ops.image.extract_patches(images, size=self.patch_size)
        patches = ops.reshape(
            patches,
            (
                batch_size,
                num_patches_h * num_patches_w,
                self.patch_size * self.patch_size * channels,
            ),
        )
        return patches

    def get_config(self):
        config = super().get_config()
        config.update({"patch_size": self.patch_size})
        return config
````

<a id="section522"></a>
## <font color="#004D7F" size=5> 5.2.2. Patch Encoder</font>

La clase `PatchEncoder` toma los parches de imagen y los proyecta en un espacio de mayor dimensión usando una capa densa (`Dense`). Luego, añade información posicional a cada parche utilizando una capa de embedding posicional (`Embedding`). Esta codificación es crucial para el funcionamiento de los Transformers, que necesitan conocer tanto el contenido de los parches como su posición en la imagen original.

<figure><center>
  <img src="../../img/vit.png" width="450" height="250" alt="Gráfica">
  <figcaption><blockquote>Arquitectura de ViT y del Trasformer Encoder. Extraída de <a href="https://arxiv.org/abs/2010.11929">artículo de ViT</a></blockquote></figcaption>
</center></figure>

La clase `PatchEncoder` es una subclase de `layers.Layer` en Keras, y se utiliza para proyectar y codificar los parches de imagen.

##### Constructor `__init__`
- `__init__` es el constructor de la clase.
- `num_patches` es el número total de parches en los que se ha dividido la imagen.
- `projection_dim` es la dimensión en la que se proyectarán los parches.
- `self.projection` es una capa `Dense` que proyecta cada parche en un espacio de mayor dimensión especificado por `projection_dim`.
- `self.position_embedding` es una capa `Embedding` que crea embeddings posicionales para cada parche, con `num_patches` como tamaño de vocabulario y `projection_dim` como dimensión de salida.

##### Método `call`
- `call` es el método que define la lógica de la capa. Es invocado cuando se pasa un tensor (en este caso, parches) a través de la capa.
- `positions` crea un tensor con una secuencia de posiciones (desde 0 hasta `num_patches - 1`), expandiendo la dimensión para que sea compatible con el batch de datos.
- `projected_patches` aplica la proyección lineal a cada parche usando la capa `Dense`, resultando en un tensor de mayor dimensión.
- `encoded` suma la proyección lineal de los parches (`projected_patches`) con los embeddings posicionales (`self.position_embedding(positions)`). Esta suma incorpora información sobre la posición de cada parche, lo que es crucial para que el Transformer entienda la disposición espacial de los parches.
- `return encoded` devuelve el tensor codificado.

##### Método `get_config`
- `get_config` es un método estándar en Keras para capas personalizadas. Permite que la configuración de la capa se pueda guardar y volver a cargar.
- `config = super().get_config()` llama al método `get_config` de la clase base para obtener la configuración básica de la capa.
- `config.update({"num_patches": self.num_patches})` añade `num_patches` a la configuración.
- `return config` devuelve la configuración completa.


````python
class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim):
        super().__init__()
        self.num_patches = num_patches
        self.projection = layers.Dense(units=projection_dim)
        self.position_embedding = layers.Embedding(
            input_dim=num_patches, output_dim=projection_dim
        )

    def call(self, patch):
        positions = ops.expand_dims(
            ops.arange(start=0, stop=self.num_patches, step=1), axis=0
        )
        projected_patches = self.projection(patch)
        encoded = projected_patches + self.position_embedding(positions)
        return encoded

    def get_config(self):
        config = super().get_config()
        config.update({"num_patches": self.num_patches})
        return config
````

<a id="section523"></a>
## <font color="#004D7F" size=5> 5.2.3. Clasificador</font>

La función `mlp` construye un MLP que transforma el tensor de entrada `x` mediante varias capas densas y de dropout. Cada capa densa aplica una transformación lineal seguida por una activación `gelu`, y cada capa de dropout aleatoriamente desactiva un porcentaje de las unidades de la capa anterior para mejorar la generalización del modelo. Esta función es útil para agregar capacidad de aprendizaje no lineal al modelo, permitiendo capturar relaciones más complejas en los datos:
- `x`: el tensor de entrada que se va a transformar.
- `hidden_units`: una lista que especifica el número de unidades (neuronas) en cada capa oculta de la red MLP.
- `dropout_rate`: la tasa de dropout que se aplicará después de cada capa densa, lo que ayuda a prevenir el sobreajuste.


<figure><center>
  <img src="../../img/vit.png" width="450" height="250" alt="Gráfica">
  <figcaption><blockquote>Arquitectura de ViT y del Trasformer Encoder. Extraída de <a href="https://arxiv.org/abs/2010.11929">artículo de ViT</a></blockquote></figcaption>
</center></figure>

````python
def mlp(x, hidden_units, dropout_rate):
    for units in hidden_units:
        x = layers.Dense(units, activation=keras.activations.gelu)(x)
        x = layers.Dropout(dropout_rate)(x)
    return x
````

<a id="section524"></a>
## <font color="#004D7F" size=5> 5.2.4. Clasificador ViT</font>

La función `create_vit_classifier` construye un clasificador Vision Transformer completo. 
1. Primero, divide la imagen de entrada en parches y codifica esos parches. 
2. Luego, pasa los parches codificados a través de múltiples capas Transformer, cada una de las cuales incluye normalización por capas, atención multi-cabeza, conexiones residuales y capas MLP. 
3. Finalmente, normaliza y aplana la representación antes de pasarla a través de una red MLP final para producir las características de salida. 

<figure><center>
  <img src="../../img/Encoder-decoder.png" width="450" height="250" alt="Gráfica">
  <figcaption><blockquote>Arquitectura de Transformer. Extraída de <a href="https://arxiv.org/abs/2010.11929">artículo de ViT</a></blockquote></figcaption>
</center></figure>


#### Entrada del modelo
- `inputs`: define la entrada del modelo con una forma especificada por `input_shape`.

#### Creación de parches
- `patches`: instancia la capa `Patches` (definida anteriormente) para dividir la imagen de entrada en parches más pequeños del tamaño especificado por `patch_size`.

#### Codificación de los parches
- `encoded_patches`: instancia la capa `PatchEncoder` (definida anteriormente) para proyectar los parches en un espacio de alta dimensión y añadir los embedding posicionales.

#### Bloques del Transformer
- Bucle a través de `transformer_layers` para crear múltiples capas del bloque Transformer.
    - `x1`: aplica la normalización por capas a los `encoded_patches`.
    - `attention_output`: aplica una capa de atención multi-cabeza.
    - `x2`: realiza una conexión residual (skip connection) sumando `attention_output` y `encoded_patches`.
    - `x3`: aplica otra normalización por capas a `x2`.
    - `x3`: pasa por un MLP usando la función `mlp` definida anteriormente.
    - `encoded_patches`: realiza otra conexión residual sumando `x3` y `x2`.

#### Representación final
- `representation`: aplica la normalización por capas, aplana el tensor y aplica dropout para la regularización.

#### Red MLP final
- `features`: aplica otra MLP a la representación final.


````python
def create_vit_classifier():
    inputs = keras.Input(shape=input_shape)
    # Augment data.
    #augmented = data_augmentation(inputs)
    # Create patches.
    patches = Patches(patch_size)(inputs)
    # Encode patches.
    encoded_patches = PatchEncoder(num_patches, projection_dim)(patches)

    # Create multiple layers of the Transformer block.
    for _ in range(transformer_layers):
        # Layer normalization 1.
        x1 = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
        # Create a multi-head attention layer.
        attention_output = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=projection_dim, dropout=0.1
        )(x1, x1)
        # Skip connection 1.
        x2 = layers.Add()([attention_output, encoded_patches])
        # Layer normalization 2.
        x3 = layers.LayerNormalization(epsilon=1e-6)(x2)
        # MLP.
        x3 = mlp(x3, hidden_units=transformer_units, dropout_rate=0.1)
        # Skip connection 2.
        encoded_patches = layers.Add()([x3, x2])

    # Create a [batch_size, projection_dim] tensor.
    representation = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
    representation = layers.Flatten()(representation)
    representation = layers.Dropout(0.5)(representation)
    # Add MLP.
    features = mlp(representation, hidden_units=mlp_head_units, dropout_rate=0.5)
    # Create the Keras model.
    model = keras.Model(inputs=inputs, outputs=features)
    return model
````

````python
vit_model = create_vit_classifier()
````

# <font color="#004D7F" size=5> 5.4. Metrics</font>

Define metrics and some hyperparameters

````python
import tensorflow.keras.backend as K

def r_square(y_true, y_pred):
    SS_res = K.sum(K.square(y_true - y_pred))
    SS_tot = K.sum(K.square(y_true - K.mean(y_true)))
    r2 = 1 - SS_res / (SS_tot + K.epsilon())
    return r2

METRICS = [
    tf.keras.metrics.MeanSquaredError(name = 'mse'),
    tf.keras.metrics.MeanAbsoluteError(name = 'mae'),
    tf.keras.metrics.RootMeanSquaredError(name = 'rmse'),
    r_square,
]
````
